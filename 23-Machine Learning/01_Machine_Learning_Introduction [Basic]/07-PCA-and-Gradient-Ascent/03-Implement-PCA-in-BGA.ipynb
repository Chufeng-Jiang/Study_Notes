## 使用梯度上升法实现PCA

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.empty((100, 2))
X[:,0] = np.random.uniform(0., 100., size=100)
X[:,1] = 0.75 * X[:,0] + 3. + np.random.normal(0, 10., size=100)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

### demean

```python
def demean(X):
    return X - np.mean(X, axis=0) # 在行方向上求均值，也就是列的均值
```

```python
X_demean = demean(X)
```

```python
plt.scatter(X_demean[:,0], X_demean[:,1])
plt.show() # （0,0）在图片的中心
```

```python
np.mean(X_demean[:,0]) # e-14 约等于 0
```

```python
np.mean(X_demean[:,1])  # e-15 约等于 0
```

### 梯度上升法

```python
def f(w, X): # 目标函数
    return np.sum((X.dot(w)**2)) / len(X)

def df_math(w, X): # 求目标函数的梯度
    return X.T.dot(X.dot(w)) * 2. / len(X)
```

```python
def df_debug(w, X, epsilon=0.0001):
    res = np.empty(len(w))
    for i in range(len(w)):
        w_1 = w.copy()
        w_1[i] += epsilon
        w_2 = w.copy()
        w_2[i] -= epsilon
        res[i] = (f(w_1, X) - f(w_2, X)) / (2 * epsilon)
    return res 
```

```python
def direction(w):# 化为单位向量
    return w / np.linalg.norm(w)

def gradient_ascent(df, X, initial_w, eta, n_iters = 1e4, epsilon=1e-8):
    
    w = direction(initial_w) 
    cur_iter = 0

    while cur_iter < n_iters:
        gradient = df(w, X)
        last_w = w
        w = w + eta * gradient
        w = direction(w) # 注意1：每次求一个单位方向
        if(abs(f(w, X) - f(last_w, X)) < epsilon):
            break
            
        cur_iter += 1

    return w
```

```python
initial_w = np.random.random(X.shape[1]) # 注意2：不能用0向量开始。 样本的特征数
initial_w
```

```python
eta = 0.001
# 注意3： 不能使用StandardScaler标准化数据
```

```python
gradient_ascent(df_debug, X_demean, initial_w, eta)
```

```python
gradient_ascent(df_math, X_demean, initial_w, eta)
```

```python
w = gradient_ascent(df_math, X_demean, initial_w, eta)
```

```python
plt.scatter(X_demean[:,0], X_demean[:,1])
plt.plot([-40, w[0]*30], [-40, w[1]*30], color='r')
plt.show()
```
