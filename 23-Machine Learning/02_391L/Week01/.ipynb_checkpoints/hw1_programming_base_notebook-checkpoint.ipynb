## 2
The goal is to build a decision tree that, based on the other features in the set, predicts
whether or not a patient has malignant cancer. So this is a classification problem. Using
`tree.DecisionTreeClassifier` and other functions in the scikit-learn library, one can build
a decision tree and calculate both its training accuracy when fitted to the entire data set as
well as its accuracy using 10-fold cross validation (which gives a better idea of true accuracy).
In this question you will need to complete two sub-components:
### (a) 
(a) Make a plot visualizing the performance of a `tree.DecisionTreeClassifier` as you
search for an optimal `max_depth` parameter. Vary the depth of your decision tree using
max depth = 1,2,. . . ,10 and record the results from the following evaluation procedures
for each setting:
* The accuracy when training and testing on the full dataset.
* 10-fold cross-validated accuracy.

Plot the results of both evaluation procedures on the same plot with evaluation scores on
the y-axis and max depth values on the x-axis. Use 10 as your random seed/state for the
decision tree and the cross-validation. Use a legend to label both evaluation procedures.


```python
x = cancer.data
y = cancer.target

# Lists to store accuracy results
full_acc = []
cross_val_acc = []

# Variables to track the best accuracies and their corresponding depths
best_full = 0
best_cross = 0
best_full_depth = []
best_cross_depth = []

# Range of max_depth values to try
max_depths = range(1, 11)

# Iterate over different max_depth values
for max_depth in max_depths:
    # Initialize Decision Tree Classifier with the current max_depth
    decision_tree = DecisionTreeClassifier(max_depth=max_depth, random_state=RANDOM_STATE)
    
    # Train and test the model on the full dataset
    full_acc_score_ = decision_tree.fit(x, y).score(x, y)
    full_acc.append(full_acc_score_)
    
    # Update the best full dataset accuracy and corresponding depths
    if full_acc_score_ > best_full:
        best_full = full_acc_score_
        best_full_depth = [max_depth]
    elif full_acc_score_ == best_full:
        best_full_depth.append(max_depth)
    
    # Perform 10-fold cross-validation
    cross_val_score_ = np.mean(cross_val_score(decision_tree, x, y, cv=10))
    cross_val_acc.append(cross_val_score_)
    
    # Update the best cross-validated accuracy and corresponding depths
    if cross_val_score_ > best_cross:
        best_cross = cross_val_score_
        best_cross_depth = [max_depth]
    elif cross_val_score_ == best_cross:
        best_cross_depth.append(max_depth)

    # Print the current max_depth, full_acc, and cross_val_acc
    print(f"max_depth: {max_depth},\t full_acc: {full_acc_score_:.4f},\t cross_val_acc: {cross_val_score_:.4f}")

# Plot the accuracy results
plt.figure(figsize=FIGSIZE)
plt.plot(max_depths, cross_val_acc, label='Cross-validated Accuracy', marker='o')
plt.plot(max_depths, full_acc, label='Full Dataset Accuracy',marker='o')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy')
plt.xticks(np.arange(1, 11, step = 1))
plt.title('Decision Tree Performance for The Breast Cancer Dataset')
plt.legend()
plt.show()

# Print the best depths for full dataset and cross-validated accuracies
print(f"Best full dataset depths: {best_full_depth}")
print(f"Best cross validated dataset depths: {best_cross_depth}")
```

## 3\. 
This question explores random forest classifiers by using scikit-learnâ€™s `ensemble.RandomForestClassifier`.
You will make two plots and answer questions about them.

### (a) 
For the first plot, use a `ensemble.RandomForestClassifier` and the best depth you
found 2(b)ii as `max_depth`. We will now find the optimal setting of a second parameter,
n estimators. Vary the number of trees in the forest via the parameter `n_estimators`
and plot its 10-fold cross-validated accuracy (use `n_estimators` = 1, 2, . . . , 20). Again,
use 10 as your random seed for your classifier and cross-validation.

```python
x = cancer.data
y = cancer.target

# Set the max_depth for the trees in the forest
max_depth = 5

# Range of n_estimators values to try
n_estimators_range = range(1, 21)

# List to store cross-validated accuracy scores
cross_validation_scores = []

# Variables to track the best accuracy and its corresponding n_estimators
best_accuracy = 0
best_n_est = []

# Iterate over different n_estimators values
for n_estimators in n_estimators_range:
    # Initialize Random Forest Classifier with the current n_estimators
    random_forest = RandomForestClassifier(max_depth=max_depth, n_estimators=n_estimators, random_state=RANDOM_STATE)
    
    # Perform 10-fold cross-validation
    cross_validation_score = np.mean(cross_val_score(random_forest, x, y, cv=KFold(n_splits=10, random_state=RANDOM_STATE, shuffle=True)))
    cross_validation_scores.append(cross_validation_score)
    
    # Update the best cross-validated accuracy and corresponding n_estimators
    if cross_validation_score > best_accuracy:
        best_accuracy = cross_validation_score
        best_n_est = [n_estimators]
    elif cross_validation_score == best_accuracy:
        best_n_est.append(n_estimators)
    
    # Print the current n_estimators and its corresponding cross-validated accuracies
    print(f"n_estimators: {n_estimators},\t cross_val_acc: {cross_validation_score:.4f}")

# Plot the cross-validated accuracy results
plt.figure(figsize = FIGSIZE)
plt.plot(n_estimators_range, cross_validation_scores, marker='o')
plt.xticks(np.arange(1, 21, step = 1))
plt.xlabel('Number of Trees')
plt.ylabel('Cross_Validated Accuracy')
plt.title('Random Forest Performance for The Breast Cancer Dataset')
plt.show()

# Print the range of n_estimators and their corresponding cross-validated accuracies
# print(f"n_estimators_range: {list(n_estimators_range)}")
# print(f"cross_validation_scores: {cross_validation_scores}")

# Print the best number of trees for the best cross-validated accuracy
print(f"Best number of trees: {best_n_est}")
```

## 4\. 
For this last question, we will explore the dependability of our estimates. 
### (a) 
Make a plot using the following procedure:
#### i. 
Using random state values from 0, 1, Â· Â· Â· , 99 calculate the 10-fold cross-validation
accuracy of different `tree.DecisionTreeClassifiers` with max depth settings from
1, 2, Â· Â· Â· , 10.
As before, you should use the same random state value for your classifier and cross-validation.
#### ii. 
Then record the best max depth settings for each random state. Be sure to check whether multiple settings
achieve the best accuracy.


Plot the counts for the best max depth settings as a bar chart with the max depth settings on the x-axis and the 'best parameter counts' on the y-axis (number of times that parameter was selected as the best max depth setting).

*Note*: this calculation might take some time. For debugging, try a smaller range of
random states.

```python
x = cancer.data
y = cancer.target

# Dictionary to store the counts of the best max_depth values
top_depths_rs = {}

# Iterate over random states from 0 to 99
for random_state_count_rs in range(100):
    best_cross_validation_score_rs = 0
    best_cross_validation_score_depth_rs = []
    
    # Iterate through all depth possibilities in a random state
    max_depths_rs = range(1, 11)
    for max_depth_count_rs in max_depths_rs:
        # Decision tree classifier with the current random_state
        decision_tree_rs = DecisionTreeClassifier(max_depth=max_depth_count_rs, random_state=random_state_count_rs)
        
        # Accuracy with 10-fold cross-validation
        cross_validation_score_rs = np.mean(cross_val_score(decision_tree_rs, x, y, cv=10))
        
        # Print the current random state, max depth, and cross-validated accuracy
        print(f"Random State: {random_state_count_rs},\t Max Depth: {max_depth_count_rs},\t Cross-Validated Accuracy: {cross_validation_score_rs:.4f}")
        
        # Check best for cross-validation
        if cross_validation_score_rs > best_cross_validation_score_rs:
            best_cross_validation_score_rs = cross_validation_score_rs
            best_cross_validation_score_depth_rs = [max_depth_count_rs]
        elif cross_validation_score_rs == best_cross_validation_score_rs:
            best_cross_validation_score_depth_rs.append(max_depth_count_rs)
    
    # Take the best depth and add to dictionary
    for depth_rs in best_cross_validation_score_depth_rs:
        top_depths_rs.setdefault(depth_rs, 0)
        top_depths_rs[depth_rs] += 1

# Extract keys and values from the dictionary
depths_rs = list(top_depths_rs.keys())
counts_rs = list(top_depths_rs.values())
```

```python
# Plot the counts of the best max_depth values
plt.figure(figsize=FIGSIZE)
plt.bar(depths_rs, counts_rs, color='green')
plt.xlabel('Max Depth')
plt.ylabel('Best Parameter Counts')
plt.xticks(np.arange(0, 11, step=1))
plt.title('Best Parameter Counts for Best Max Depth')

# Add labels to the top of each bar
for i, count_rs in enumerate(counts_rs):
    plt.text(depths_rs[i], count_rs, str(count_rs), ha='center', va='bottom')

plt.show()

# Print the dictionary with counts for best max_depth values
print("Counts for Best Max Depths:", top_depths_rs)
```
